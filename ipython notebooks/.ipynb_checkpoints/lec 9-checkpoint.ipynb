### slicing it [row,col] so upto 2 rows(exclusive) and from 1 rows onwards(inclusive)

```python
arr2d = np.zeros([10,10])
```

```python
arr2d
```

```python
arr_length = arr2d.shape[1]
```

```python
arr_length
```

```python
for i  in range(arr_length):
    arr2d[i] = i
```

```python
arr2d
```

```python
arr2d[[2,4,6,8]]
```

###fancy indexing

```python
arr2d[[i for i in range(10) if i%2 ==0 ]]
```
